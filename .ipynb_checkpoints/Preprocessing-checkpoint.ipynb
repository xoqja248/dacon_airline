```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
#dataset 불러오기 
train = pd.read_csv("airline_dataset/train.csv")
test = pd.read_csv("airline_dataset/test.csv")

#dataset 확인
print(train.columns)
train.head()
```

```python
train['Ease of Online booking'].unique()
```

```python
# id column 삭제
train.drop('id',axis=1,inplace =True)
test.drop('id',axis=1, inplace =True)


# Gender one_hot encoding
train = pd.get_dummies(train,drop_first=True,columns = ['Gender',
                                                         'Customer Type',
                                                         'Type of Travel',
                                                         'Class',
                                                         'Seat comfort',
                                                         'Food and drink',
                                                         'Inflight entertainment',
                                                         'Leg room service',
                                                         'Online boarding',
                                                         'Ease of Online booking',
                                                         'Inflight wifi service'])

test = pd.get_dummies(test,drop_first=True,columns = ['Gender',
                                                        'Customer Type',
                                                        'Type of Travel',
                                                        'Class',
                                                        'Seat comfort',
                                                        'Food and drink',
                                                        'Inflight entertainment',
                                                        'Leg room service'])


# Label 0 데이터가 train에만 존재하기 때문에 drop_first = False를 주고 Label 1 데이터를 살려 columns 수와 순서 일치시킴

test = pd.get_dummies(test,drop_first=False,columns = [ 'Online boarding',
                                                        'Ease of Online booking'])

test = pd.get_dummies(test,drop_first=True,columns = ['Inflight wifi service'])

# Departure/Arrival time convenient 삭제 무응답 경우의 수가 많고 target과의 상관관계도 매우 낮음)

train.drop('Departure/Arrival time convenient',axis = 1,inplace = True)
test.drop('Departure/Arrival time convenient',axis = 1,inplace = True)

# Departure Delay in Minutes, Arrival Delay in Minutes, log transform

train['Departure Delay in Minutes'] = np.log1p(train['Departure Delay in Minutes'])
test['Departure Delay in Minutes'] = np.log1p(test['Departure Delay in Minutes'])

train['Arrival Delay in Minutes'] = np.log1p(train['Arrival Delay in Minutes'])
test['Arrival Delay in Minutes'] = np.log1p(test['Arrival Delay in Minutes'])
```

```python
# column이 하나 더 많은것은 target data 때문임

train.shape
```

```python
test.shape
```

```python
X_train = train.drop('target',axis=1)
y_train = train['target']
```

```python
#MinMaxscaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
scaler.partial_fit(test)
```

```python
X_train = pd.DataFrame(scaler.transform(X_train),columns = X_train.columns)
test = pd.DataFrame(scaler.transform(test), columns = test.columns)
```
